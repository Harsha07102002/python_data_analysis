```python
so = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.barplot(x='State',y='Orders',data = so)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=so,x='State',y='Orders')
```

```python
sc = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True).head(10)
sns.barplot(x='State',y='Amount',data = sc)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sc,x='State',y='Amount')
```
